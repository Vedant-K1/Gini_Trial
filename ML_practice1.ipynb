```python
import numpy as np
import pandas as pd
data = pd.read_csv("EXP3_1.csv")
data
```

```python
n = data.shape[0]
ncol = data.shape[1]
decision = data.iloc[:,-1:]
data3= data.iloc[:,1:ncol-1]
```

```python
listInc = list(data3[data3.columns[0]].unique())
listAge = list(data3[data3.columns[1]].unique())
```

```python
yes=0
rented=0
w_avg=0
for i in listInc:
  for j in range(n):
    if(data3["INCOME"][j]==i):
      if(decision["OWNHOUSE"][j]=="YES"):
        yes=yes+1
      else:
        rented=rented+1
  g_index = 1 - (yes/(yes+rented))**2 - (rented/(yes+rented))**2
  w_avg = w_avg + g_index*((yes+rented)/n)
  g_index=0
  yes=0
  rented=0
  
```

```python
yes=0
rented=0
w_avg2=0
for i in listAge:
  for j in range(n):
    if(data3["AGE"][j]==i):
      if(decision["OWNHOUSE"][j]=="YES"):
        yes=yes+1
      else:
        rented=rented+1
  g_index = 1 - (yes/(yes+rented))**2 - (rented/(yes+rented))**2
  w_avg2 = w_avg2 + g_index*((yes+rented)/n)
  g_index=0
  yes=0
  rented=0
w_avg2
```
